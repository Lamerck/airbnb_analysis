```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
```

```python
plt.plot(x, y, color='red')
```

```python
plt.scatter(x, y, c='red')
```

```python
x1 = [2, 8, 39, 25, 69, 45, 100, 121]
y1 = [49, 54, 78, 84, 73, 96, 90, 105]
colors = ['green', 'purple', 'yellow', 'blue', 'pink', 'brown', 'black', 'aqua']
sizes = [10, 20, 50, 100, 150, 200, 500, 750]
plt.scatter(x1, y1, marker='o', c=colors, s=sizes)
```

```python
# large dataset
x3 = np.random.rand(100)
y3 = np.random.rand(100)
colors = np.random.rand(100)
plt.scatter(x3, y3, c=colors, alpha=0.5, cmap='viridis')
```
