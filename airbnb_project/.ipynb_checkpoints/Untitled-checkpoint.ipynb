```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
```

```python
plt.plot(x, y, 'o', color='red')
plt.ylim(-1.2, 1.2)
```

```python
# plt.scatter
x1 = [0, 2, 4, 6, 8, 10]
y1 = [0, 4, 8, 36, 64, 100]
colors = ['red', 'green', 'blue', 'black', 'purple', 'yellow']
sizes = [10, 20, 50, 80, 200, 500]
plt.scatter(x1, y1, marker='o', c=colors, s=sizes)
```

```python
# plt.scatter
x2 = np.linspace(0, 12, 6)
y2 = np.sin(x2)
colors = ['red', 'green', 'blue', 'black', 'purple', 'yellow']
sizes = [10, 20, 50, 80, 200, 500]
plt.scatter(x2, y2, marker='o', c=colors, s=sizes)
```

```python
# for larger datasets
x3 = np.random.rand(100)
y3 = np.random.rand(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

plt.scatter(x3, y3, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.ylim(-0.1, 1.1)
```

```python
# less beautiful but more complex
x4 = [1, 2, 3, 4, 5]
y4 = [10, 20, 25, 30, 35]
x5 = [1, 2, 3, 4, 5]
y5 = [40, 30, 25, 20, 10]

plt.scatter(x4, y4, color='blue', label='Dataset 1' )
plt.scatter(x5, y5, color='red', label='Dataset 2')
```
